```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
###    Draw a Line in a diagram from position(0,10) to position (5,150):
xpoints = np.array([1,2])
ypoints = np.array([5,10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1,10])
ypoints = np.array([20,200])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1,6])
ypoints = np.array([4,5])

plt.plot(xpoints, ypoints, "*")
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, "*""b")
plt.show()
```

```python
xpoints = np.array([3, 8, 1, 12])
ypoints = np.array([5, 6, 7, 12])

plt.plot(xpoints, ypoints, marker= 'o', ms= 50, color='red')
plt.show()
```

```python
xpoints = np.array([3, 8, 1, 12])
ypoints = np.array([5, 6, 7, 12])

plt.plot(xpoints, ypoints, marker= 'o', ms= 50, mec='b', color='pink')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc= 'r', mec= 'k')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 200, mfc= 'k', mec= 'y')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, linestyle='dashed', linewidth=1, marker='o', mec='hotpink', ms=20, mfc='blue')
plt.show()
```

```python
xpoints = np.array([10, 20, 30])
ypoints = np.array([1, 2, 6])

plt.plot(xpoints, ypoints, marker="*", color='red', mfc='g', mec='r',ms =20, linestyle='dashed')
```

```python
x = np.array([10, 20, 30, 40 , 50, 60, 70, 80, 90])
y = np.array([5, 15, 25, 35, 45, 55, 65, 75, 85])

plt.plot(x,y, color='red', linewidth=1)

plt.xlabel("Average Pulse", fontsize=10, color='red', loc='right')
plt.ylabel("Calorie Burnage", fontsize=10, color='green')
plt.title('sample', fontsize=50, color='blue', loc='left')
plt.show()
```

```python
x = np.array([10, 20, 30, 40 , 50, 60, 70, 80, 90])
y = np.array([5, 15, 25, 35, 45, 55, 65, 75, 85])

plt.title("Sports Watch Data")
plt.xlabel("Average Pluse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y, color="orange")
plt.grid(color="green", linestyle="dotted")  ## We also specifies the axis for grid as axis="x"
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pluse")
plt.ylabel("Calorie Burnage")

plt.plot(x,y)

plt.grid(color = 'hotpink', linestyle='--', linewidth = 1.5)

plt.show()
```

```python
# plot 1:
x = np.array([0,1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2,1,1)
plt.plot(x,y)

# plot 2:
x = np.array([0,1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2,1,2)
plt.plot(x,y)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,10,15,20,25])
plt.scatter(x,y,color='red')
plt.xlabel("id", color='green')
plt.ylabel("marks")
plt.title('Reln b/w id and marks')
plt.show()
```

```python
x =np.array([1,2,3,4,5])
y =np.array([6,7,8,9,10])
colors=(['red', 'yellow', 'black', 'blue', 'green'])
size=([10,200,30,40,50])
plt.scatter(x,y,c=colors,s=size)
plt.show()
```

```python
x =np.array(['A', 'B', 'C', 'D', 'E'])
y =np.array([6,7,8,9,10])

### If we want bar in horizontal direction use plt.barh, width=0.1, height=0.5...
plt.bar(x,y, color='black')
plt.show()
```

```python
y = np.array([10, 135, 25, 15, 19])

plt.pie(y)
plt.show()
```

```python
y = np.array([10, 135, 25, 15])
mylabels= ['Cherries', 'Apples', 'Bananas', 'Dates']

plt.pie(y, labels = mylabels, startangle=180)
plt.show()
```

```python
y = np.array([10, 135, 25, 15])
mylabels= ['Cherries', 'Apples', 'Bananas', 'Dates']
myexplode= [0, 0.95, 0.8, 0.2]

plt.pie(y, labels = mylabels, explode=myexplode)
plt.show()
```

```python
y =np.array([35, 25, 25, 15])
myLabels = ["Apples", 'Bananas', 'Cherries', 'Dates']

plt.pie(y)
plt.legend(labels=myLabels)
plt.show()
```
